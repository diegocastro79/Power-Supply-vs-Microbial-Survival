# The current notebook generates all the plots appearing in the main manuscript of the publication: Diego et al., "On how the power supply shapes microbial survival", accepted for publication in Mathematical Biosciences 
## The code presented in this notebook depends on a customized functions contained in the notebooks Plots_Class and Utils, included in the current repository

## Importing modules

```python
import numpy as np
import time 
import Plots_Class as pc # customized class containing functions to generate the plots
```

## Initializing the customized module containing the functions for plot generation

```python
x = pc.Plots()
```

## * Generating the plots $H_B$ vs $H_P$

### Setting the optional variables 

#### Witdh of the distribution for the model parameters, $\sigma$

```python
sigma = np.array([0.001, 0.01, 0.1, 1])
```

#### Setting which parameter(s) is (are) non-trivially distributed across habitats

```python
subst_flat = False 
# If False: The values for the input substrate concentrations follow a Gaussian distribution
# If True: All the specices share the same value for the input substrate concentration 
```

```python
gibbs_flat = False 
# If False: The values for the Gibbs energy of reactions follow a Gaussian distribution
# If True: All the specices share the same value for the Gibbs energy of reaction
```

```python
tradeoff = True 
# If True: An efficiency-cost trade-off is considered. 
           # That is: both uptake rates, $r_i$, and maintenance powers, $P_{0i}$
           # follow Gaussian distributions
# If False: All the species share the same values for the uptake rates and the maintenance powers
```

```python
x.HBvsHP(sigma=sigma,
        subst_flat=subst_flat,
        gibbs_flat=gibbs_flat,
        tradeoff=tradeoff)
```

#### The color gradient for the graphics above shows the number of habitats from $N=10$ (red) to $N=1000$ (green)
